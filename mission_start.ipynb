# 1. Defining Variables
# Integer (Whole Numbers)

```python
student_id = 101
print(f"student ID: {student_id}")
```

# Float (Decimal Number)

```python
cgpa = 9.8
print(f"cgpa: {cgpa}")
```

# String (Text)

```python
Student_name = "Arjun"
print(f"Student Name: {Student_name}")
```

# Boolean (True/false - Logic)

```python
is_registered = True
print(f"Is registered: {is_registered}")
```

# 2. Checking Data Types using type() function

```python
print(type(student_id))
print(type(cgpa))
print(type(student_name))
print(type(is_registered))
```

# 3. Dynamic Typing

```python
data = 100
print(f"Data is now: {data}, type: {type(data)}")

data = "now I am a string"
print(f"data is now: {data}, type: {type(data)}")
```
