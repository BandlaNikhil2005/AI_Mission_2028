# Module 2: Python Lists (Data Structures)

# 1. LIST CREATION

```python
response_time = [120, 45, 230, 15, 90]

print(f"server times: {response_time}")
```

# 2. INDEXING (Accessing Data)

```python
first_request = response_time[0]
print(f"First Response Time: {first_request}")
```

```python
third_request = response_time[2]
print(f"third response time: {third_request}")
```

```python
last_request = response_time[-1]
print(f"last response time: {last_request}")
```

# 3. ANALYSIS FUNCTIONS

```python
fastest = min(response_time)
print(f"fastest response: {fastest} ms")
```

```python
slowest = max(response_time)
print(f"slowest response: {slowest} ms")
```

```python
total_requests = len(response_time)
print(f"total requests: {total_requests}")
```

```python
avg_time = sum(response_time) / len(response_time)
print(f"Average response time: {avg_time} ms")
```

# Module 3: Python Dictionaries (Real-World Data)

# 1. CREATION (USE CURLY BRACKETS{})

```python
student_record = {
    "name" : "Arjun Vasudev",
    "roll_number" : "2028MCA01",
    "skills": ["python", "SQL", "AI"],
    "cgpa": "9.8",
    "is_placed": False
}

print(f"student record: {student_record}")
```

# 2. ACCESSING DATA (No Index, Only Key)

```python
print(f"student_name: {student_record['name']}")
```

```python
print(f"skills: {student_record['skills']}")
```

# 3. MODIFYING DATA 

```python
student_record['is_placed'] = True
student_record['skills'].append("deep learning")

print(f"updated record: {student_record}")
```

# 4. KEYS & VALUES 

```python
print(f"labels: {student_record.keys()}")
print(f"values: {list(student_record.values())}")
```
